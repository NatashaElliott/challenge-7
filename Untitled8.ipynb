```python
from pathlib import Path
import pandas as pd
```

```python
census_2018 = Path("Documents/censusdata2018.csv")
ambulance_2018 = Path("Documents/ambulance2018.csv")
```

```python
census_data_2018 = pd.read_csv(census_2018)
ambulance_data_2018 = pd.read_csv(ambulance_2018)
```

```python
census_data_2018_vic = census_data_2018[census_data_2018["S/T name"] == "Victoria"]
census_data_2018_vic = census_data_2018_vic.iloc[:,[9,28]]
```

```python
merged_df = pd.merge(ambulance_data_2018, census_data_2018_vic, left_on="UCL Name", right_on="SA2 name", how="left")
missing_ucls = merged_df[pd.isnull(merged_df["SA2 name"])][merged_df.columns]
```
